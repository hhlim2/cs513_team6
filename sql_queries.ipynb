### Q1: When did different types of coffee drinks (dishes) begin appearing on menus and become more prevalent?

```python
%%sql
select name, first_appeared_fixed, times_appeared from DishCleaned group by first_appeared_fixed
```

### Q2: Based on the dataset, has coffee always been on restaurants menus since the earliest record available?

```python
# find minimum first_appeared date in Dish dataset, compare to minimum first_appeared date in DishCleaned dataset - filtering for invalid dates

%%sql
select min(first_appeared) from Dish where first_appeared > 1000
```

```python
%%sql
select min(first_appeared_fixed) from DishCleaned
```

The minimum first_appeared value for a dish in the Dish dataset is 1851, in the DishCleaned dataset with only coffee dishes it is 1882
